```
pal = ["#f2fa00", "#f20521", "#05f2f2", "#0528f2", "#00d11f", "#007d3c"]
```

```
HP_DF = pd.DataFrame(housePiano)
HG_DF = pd.DataFrame(houseGuitar)
HD_DF = pd.DataFrame(houseDrums)
HB_DF = pd.DataFrame(houseBass)
joinedH = pd.concat([HP_DF,HG_DF,HD_DF,HB_DF],  axis=1)
joinedH.columns = ['Piano', 'Guitar',  'Drums',  'Bass']
fig, ax = plt.subplots()
fig.set_size_inches(11.7, 8.27)
ax = sns.boxplot(ax=ax, x="variable", y="value", data=pd.melt(joinedH), palette=sns.color_palette(pal))
sns.swarmplot(ax=ax, data=joinedH, size=2.5, color=".55", linewidth=0)
ax.set(xlabel='Instrument', ylabel='Loss', title="House")
plt.show()
```
